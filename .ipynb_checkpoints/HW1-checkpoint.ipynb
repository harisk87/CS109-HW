Transform the data set to have years as the rows and countries as the columns. Show the head of this data set when it is loaded. 

```python
# List of indices (do it this way to avoid GDP per capita appearing in the dataframe), which are the country names
tmpidx = [i for i in df2['GDP per capita']]
# Get the transpose but remove the first row which thinks all the country names are data points
df2t = df2.transpose()[1:]
# Set the column titles to be the indices from the original dataframe
df2t.columns = tmpidx
# Remove all columns with NaN elements
df2t = df2t.dropna(axis=1,how='any')
df2 = df2t
```

#### Problem 2(b)

Graphically display the distribution of income per person across all countries in the world for any given year (e.g. 2000).  What kind of plot would be best? 

```python
def plotForYear(year):

    df2_single_year= df2[df2.index==year].values[0]
    plt.hist(np.log10(df2_single_year.astype(int)),bins=50)
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    plt.xlabel('log10(Income per person)',fontsize=40)
    plt.ylabel('Frequency', fontsize=40)
    plt.title('Distribution of incomes in the year ' + str(year),fontsize=30)
```

```python
plotForYear(1900)
```

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
def mergeByYear(year):
    #Create a new dataframe with the values of the GDP for all countries that year
    df3 = pd.DataFrame(df2[df2.index==year].values[0])
    
    # Set the column name to be Income
    df3.columns = ['Income']
    
    # In this datafarame, add a new column and populate it with the country names from df2
    df3['Country'] =  [i for i in df2[df2.index==year].columns]
    
    # Merge this with the regions dataframe, only on the countries that have a region attached to them
    df3 = pd.merge(df1,df3,on=['Country'],how='left')
    
    
    return df3
```

```python
mergeByYear(1800)
```

#### Problem 3(a)

Assume you have two list of numbers, X and Y, with distribution approximately normal. X and Y have standard deviation equal to 1, but the average of X is different from the average of Y.  If the difference in the average of X and the average of Y is larger than 0, how does the proportion of X > a compare to the proportion of Y > a? 

Write a function that analytically calculates the ratio of these two proportions: Pr(X > a)/Pr(Y > a) as function of the difference in the average of X and the average of Y. 

**Hint**: Use the `scipy.stats` module for useful functions related to a normal random variable such as the probability density function, cumulative distribution function and survival function. 

**Update**: Assume Y is normally distributed with mean equal to 0. 

Show the curve for different values of a (a = 2,3,4 and 5).

```python
"""
Function
--------
ratioNormals

Return ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a) as function of 
    the difference in the average of X 
    and the average of Y. 

Parameters
----------
diff : difference in the average of X 
    and the average of Y. 
a : cutoff value

Returns
-------
Returns ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a)
    
Example
-------
>>> ratioNormals(diff = 1, a = 2)
"""
def ratioNormals(diff,a):
    Pyga = sp.norm.sf(a,0,1)
    Pxga = sp.norm.sf(a,diff,1)
    return Pxga/Pyga
```

```python
#your code here

# let diff range from 0 to 5 
diffs = np.linspace(0, 5, 50)
a_values = range(2,6)

# Plot separate curves for 
# Pr(X > a) / Pr(Y > a) as a function of diff
# for all given values of a
plt.figure(figsize=(8,5));
for a in a_values:
    ratios = [ratioNormals(diff, a) for diff in diffs]
    plt.plot(diffs, ratios)
    
# Labels
plt.legend(["a={}".format(a) for a in a_values], loc=0);
plt.xlabel('Diff');
plt.ylabel('Pr(X>a) / Pr(Y>a)');
plt.title('Ratio of Pr(X > a) to Pr(Y > a) as a Function of Diff');

# Using a log scale so you can actually see the curves
plt.yscale('log')
```
